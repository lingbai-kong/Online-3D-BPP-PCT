```python
import onnx

onnx_model = onnx.load("DRL_GAT-sim.onnx")
graph = onnx_model.graph
node  = graph.node

for i in range(len(node)):
    print(node[i])
```

```python
for i in range(len(node)):
    if node[i].op_type == 'Multinomial':
        print(i)#105
        print(node[i])
```

```python
old_scale_node = node[339]
new_scale_node = onnx.helper.make_node(
    "Constant",
    inputs=[],
    outputs=['449'],
    value=onnx.helper.make_tensor('value', onnx.TensorProto.FLOAT, [4], [1, 1, 1.81, 1.81])
)  # 新建新节点
graph.node.remove(old_scale_node)  # 删除旧节点
graph.node.insert(339, new_scale_node)  # 插入新节点
```
